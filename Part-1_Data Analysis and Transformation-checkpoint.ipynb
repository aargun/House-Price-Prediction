## Missing values
Let's calculate _percentage_ of missing values per variable. And as per the percentage handle each variable as required.

```python
def print_missing_percentage(X):
    """
        Prints the missing values.
        INPUT: It takes Pandas Series as input.
    """
    _percentage = X[X.isnull()].shape[0]/X.shape[0]
    if _percentage > 0.0:
        print(f"'{X.name}' missing percentage: {_percentage * 100}%")
    return False

data.apply(print_missing_percentage, axis=0);
```

As per the above result, since variables ['Alley', 'PoolQC', 'Fence', 'MiscFeature'] have so many missing values, it doesn't add anything to analysis. Thus, removing these variables

```python
# Dropping columns 'Alley', 'PoolQC', 'Fence', 'MiscFeature'
data = data.drop(columns=['Alley', 'PoolQC', 'Fence', 'MiscFeature'], axis=1)

# Shape
print(data.shape)
```

### Numerical Variable
Lets identify numerical variable and fill the missing value.

```python
# list of numerical variables
num_vars = [var for var in data.columns if data[var].dtypes != 'O']

print('Number of numerical variables: ', len(num_vars))

# visualise the numerical variables
data[num_vars].head()
```

```python
data[num_vars].apply(print_missing_percentage, axis=0);
```

```python
print(data[['LotFrontage', 'MasVnrArea', 'GarageYrBlt']].mode())
data[['LotFrontage', 'MasVnrArea', 'GarageYrBlt']].describe()
```

As, we see from above analysis that only __3 Numeric Variables__ have missing value. Best would be to use __mode__ value to replace the missing values.

```python
data[['LotFrontage', 'MasVnrArea', 'GarageYrBlt']]=data[['LotFrontage', 'MasVnrArea', 'GarageYrBlt']].fillna(data.mode().iloc[0])
data[num_vars].apply(print_missing_percentage, axis=0);
```

## Outliers
Let's detect the outliers in __continuous variables__ and __rare labels__ in categorical varibales.

First let's identify continuous varibales

```python
# list of variables that contain year information
year_vars = [var for var in num_vars if 'Yr' in var or 'Year' in var]

print('Number of years variables: ', len(year_vars))

#  list of discrete variables, assuming 20 as threshold
discrete_vars = [var for var in num_vars if len(data[var].unique())<20 and var not in year_vars+['Id']]

print('Number of discrete variables: ', len(discrete_vars))

# list of continuous variables
cont_vars = [var for var in num_vars if var not in discrete_vars+year_vars+['Id']]

print('Number of continuous variables: ', len(cont_vars))

data[cont_vars].head()
```

```python
# let's make boxplots to visualise outliers in the continuous variables 

def find_outliers(df, var):
    df = df.copy()
    
    #log does not take negative values, so let's be careful and skip those variables
    if 0 in data[var].unique():
        pass
    else:
        df[var] = np.log(df[var])
        df.boxplot(column=var)
        plt.title(var)
        plt.ylabel(var)
        plt.show()
    
for var in cont_vars:
    find_outliers(data, var)
```

As it can be seen in the __box plot__ that, there are lot of outliers in continuous varibale and __outliers__ does affect __linear regression__, one needs to handle them. Removing them can be tricky as it results in loss of information. Let __apply transformation__ and calculate accuracy too see if removing outliers would improve the accuracy or not.

#### Rare Labels
Let's evaluate how many different categories are present in each of the variables.

```python
for var in cat_vars:
    print(var, len(data[var].unique()), ' categories')
```

```python
def analyse_rare_labels(df, var, rare_perc):
    df = df.copy()
    tmp = df.groupby(var)['SalePrice'].count() / len(df)
    return tmp[tmp<rare_perc]

for var in cat_vars:
    print(analyse_rare_labels(data, var, 0.01))
    print()
```

So, all categories have labels which are in less than 1% of data. __Thus, would be removing them by replacing them with 'Rare'.__

```python
def find_frequent_labels(df, var, rare_perc):
    # finds the labels that are shared by more than a certain % of the houses in the dataset
    df = df.copy()
    tmp = df.groupby(var)['SalePrice'].count() / len(df)
    return tmp[tmp>rare_perc].index

for var in cat_vars:
    frequent_ls = find_frequent_labels(data, var, 0.01)
    data[var] = np.where(data[var].isin(frequent_ls), data[var], 'Rare')
print('Rare Labels Removed')
```

```python
# let's explore the relationship between the year variables and the house price in a bit of more details
def analyse_year_vars(df, var):
    df = df.copy()
    
    # capture difference between year variable and year the house was sold
    df[var] = df['YrSold'] - df[var]
    
    plt.scatter(df[var], df['SalePrice'])
    plt.ylabel('SalePrice')
    plt.xlabel(var)
    plt.show()
    
for var in year_vars:
    if var !='YrSold':
        analyse_year_vars(data, var)
```

## Transformations
In this section, we would transform some of the variables as per observations above.

### Year Data
We would transform the year data to difference between _YrSold_ and others

```python
def elapsed_years(df, var):
    
    # capture difference between year variable and year the house was sold
    df[var] = df['YrSold'] - df[var]
    return df

for var in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    data = elapsed_years(data, var)
    
print('Year Data Transformed')
```

### Numerical Data
We will log transform the numerical variables that do not contain zeros in order to get a more Gaussian-like distribution.

```python
for var in ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']:
    data[var] = np.log(data[var])
```

## Categorical Data
We will encode categorical variables here. But we need to first split the data into train and test to aviod overfitting.

```python
# to divide train and test set
from sklearn.model_selection import train_test_split

# Let's separate into train and test set

X_train, X_test, y_train, y_test = train_test_split(data, data.SalePrice,
                                                    test_size=0.1,
                                                    random_state=0) # we are setting the seed here
X_train.shape, X_test.shape
```

```python
# this function will assign discrete values to the strings of the variables, 
# so that the smaller value corresponds to the smaller mean of target

def replace_categories(train, test, var, target):
    ordered_labels = train.groupby([var])[target].mean().sort_values().index
    ordinal_label = {k:i for i, k in enumerate(ordered_labels, 0)} 
    train[var] = train[var].map(ordinal_label)
    test[var] = test[var].map(ordinal_label)

for var in cat_vars:
    replace_categories(X_train, X_test, var, 'SalePrice')

print('Categorical values transformed')
```

## Feature Scaling
For use in linear models, features need to be either scaled or normalised. In the next section, I will scale features between the min and max values:

```python
# feature scaling
from sklearn.preprocessing import MinMaxScaler
```

```python
train_vars = [var for var in X_train.columns if var not in ['Id', 'SalePrice']]
len(train_vars)
```

```python
# fit scaler
scaler = MinMaxScaler() # create an instance
scaler.fit(X_train[train_vars]) #  fit  the scaler to the train set for later use

# transform the train and test set, and add on the Id and SalePrice variables
X_train = pd.concat([X_train[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(X_train[train_vars]), columns=train_vars)],
                    axis=1)

X_test = pd.concat([X_test[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(X_test[train_vars]), columns=train_vars)],
                    axis=1)

print('Train Test Scaled!')
```
